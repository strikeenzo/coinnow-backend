```python
import math
import random

def generate_offset(change_amount_min, change_amount_max, origin_total, current_total, index, total_res):
    if total_res > 0:
        change_amount_min += total_res
        if change_amount_min > 0:
            change_amount_min = -100

    if total_res < 0:
        change_amount_max += total_res
        if change_amount_max < 0:
            change_amount_max = 100
            
    return change_amount_min, change_amount_max
```

```python
def generate_random_amount(total, change_amount, offset_min, offset_max, origin_total, current_total, index, total_res):
    change_amount_min, change_amount_max = generate_offset(-change_amount + offset_min, change_amount + offset_max, origin_total, current_total, index, total_res)
    print(change_amount_min, change_amount_max)
    return total + random.randint(change_amount_min, change_amount_max)

def predict(marketplace):
    total_res, seg_count, seg_index, offset_min, offset_max = 0, 4, 0, 0, 0
    for i in range(len(marketplace)):
        next_total_amount = generate_random_amount(marketplace.iloc[i].total, marketplace.iloc[i].total_change_amount, offset_min, offset_min, marketplace.iloc[i].origin_total, marketplace.iloc[i].total, i, total_res)
        
#         profit = marketplace.iloc[i].origin_total - next_price * marketplace.iloc[i].quantity 
        next_price = next_total_amount // marketplace.iloc[i].quantity
        
        marketplace.iloc[i, marketplace.columns.get_loc('next_price')] = next_price
#         marketplace.iloc[i].next_price = next_total_amount // marketplace.iloc[i].quantity
        marketplace.iloc[i, marketplace.columns.get_loc('next_total_amount')] = marketplace.iloc[i].next_price * marketplace.iloc[i].quantity
#         marketplace.iloc[i].next_total_amount = marketplace.iloc[i].next_price * marketplace.iloc[i].quantity
        total_res += marketplace.iloc[i].origin_total - marketplace.iloc[i].next_total_amount
    return total_res, marketplace
```
